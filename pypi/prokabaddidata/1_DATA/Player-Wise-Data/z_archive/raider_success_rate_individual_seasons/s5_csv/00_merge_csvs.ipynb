```python
import pandas as pd
import os

```

```python
files = []
for file in os.listdir("."):
    if file.endswith("conv.csv"):
        files.append(file)

len(files), files

```

```python
for file in files:
    print(file.split("s5")[0].strip())
```

```python
df_list = []
seas = "s5"

for file in files:
    df = pd.read_csv(f"./{file}")
    df['team'] = file.split("s5")[0].strip()
    df['season'] = "PKL-S" + seas[-1]
    df_list.append(df)
    print(df.head())
    print("\n\n")


```

```python
merged_df = pd.concat(df_list, axis=0, ignore_index=True)
```
