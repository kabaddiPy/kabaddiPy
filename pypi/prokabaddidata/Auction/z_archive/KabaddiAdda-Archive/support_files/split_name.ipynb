```python
import pandas as pd

```

```python
df = pd.read_csv("./kabaddiadda_auctiondata_FINAL.csv")
```

```python
df.head()
```

```python
capital_names = []
for index, row in df.iterrows():
    s = row['Player_Name']
    l = s.split(' ')
    
    l = [i.capitalize() for i in l]
    print(l)
    print("----")

    print(" ".join(l))

    capital_names.append(" ".join(l))

    # row['Player_Name'] = " ".join(l)

```

```python
df_name = pd.DataFrame(capital_names, columns=["Player_Name_Final"])
```

```python
import pandas as pd

# Concatenate the two dataframes along the columns axis
combined_df = pd.concat([df, df_name['Player_Name_Final']], axis=1)

# Print the combined dataframe
combined_df
```

```python
combined_df.drop(['Unnamed: 0', 'Player_Name'], inplace=True, axis=1)
```

```python
combined_df.head()
```

```python
combined_df.columns
```

```python
combined_df = combined_df[['Player_Name_Final', 'Tournament', 'Team', 'Price', 'Status']]
combined_df.head()
```

```python
# Rename a single column
combined_df.rename(columns={'Player_Name_Final': 'Player_Name'}, inplace=True)

```
